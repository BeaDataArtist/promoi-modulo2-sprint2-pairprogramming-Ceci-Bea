3. Usando "information_schema", extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database = "northwind")
mycursor = cnx.cursor()
mycursor.execute("SHOW COLUMNS FROM customers")
myresult = mycursor.fetchall()
column_names = []

# Usar un bucle for para agregar solo los nombres de las columnas a la lista
for column in myresult:
    column_names.append(column[0])
for x in column_names:
    print(x)

```

4. Seleccionar todo de la tabla "customers" donde "country" sea "USA" o "Germany". Almacenar los resultados en una variable.

```python
# Crear el cursor
mycursor = cnx.cursor()

# Definir la consulta y los valores
query = "SELECT * FROM customers WHERE country = %s OR country = %s"
values = ('USA', 'Germany')

# Ejecutar la consulta
mycursor.execute(query, values)

# Recuperar los resultados y almacenarlos en una variable
resultados = mycursor.fetchall()

# Cerrar el cursor
mycursor.close()

# (Opcional) Mostrar los resultados
for resultado in resultados:
    print(resultado)

```
